# This is generator of dummy data for training Feature Selection

### Imports

```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_regression
```

### Constants

```python
NR_FEATURES = 300
NR_SAMPLES = 1000
RANDOM_STATE = 42
CONSTANT_NUM = 88
OUTPUT_FILE = 'dataset_1.csv'
```

```python
# create the X and y
X, y = make_regression(n_samples=NR_SAMPLES, n_features=NR_FEATURES,
                       random_state= RANDOM_STATE, noise=1)
```

```python
# header_name = []
# for n in range (NR_FEATURES):
#     header_name.append('var_' + str(n).zfill(3))

header_name = ['var_' + str(n).zfill(3) for n in range(NR_FEATURES)]
```

```python
data = pd.DataFrame(X, columns=header_name)
data['target'] = y

data.head()
```

```python
# make 3 columns of constants
list_columns_cte = ['var_002', 'var_003', 'var_005']

data[list_columns_cte] = CONSTANT_NUM
data.head()
```

```python
# copy value of one column in others columns
col_in = 'var_006'
col_out = ['var_007', 'var_008']

for col in col_out:
    data[col] = data[col_in]

data.head()
```

```python
# make a quasi-constant column
col_quasi = list_columns_cte[0]

# change just the three first row
data.loc[0:2, col_quasi] = CONSTANT_NUM - 0.01
data.head()
```

### Save in csv file

```python
data.to_csv(OUTPUT_FILE, index=False)
```
